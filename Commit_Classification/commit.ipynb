```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
# from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
# from statistics import mean
import numpy
from sklearn import metrics
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
```

```python
#currently_working
#https://towardsdatascience.com/a-practical-guide-to-implementing-a-random-forest-classifier-in-python-979988d8a263


def read_csv_dataset(csv_file, csv_delimiter):
  global  rows, cols, commits, X, y
  commits = []
  csv_contents = pd.read_csv(csv_file, delimiter=csv_delimiter)

  X = csv_contents.drop(['commitId', 'project', 'comment', 'label'], axis = 1)
  y = csv_contents['label']
  

  cv = ShuffleSplit(n_splits=10, test_size=0.15, random_state=42)

  clf=RandomForestClassifier()
  #clf.fit(training, training_labels)
  # scores = cross_val_score(clf, X, y, cv=cv)
  for (train, test), i in zip(cv.split(X, y), range(10)):
    # print(1)
    # print(train)
    # print(2)
    # print(y.iloc[test])
    clf.fit(X.iloc[train], y.iloc[train])
    y_actual = y.iloc[test].values
    # print(i, 10000000000)
    #for row in X.iloc[test]:
    y_pred = clf.predict(X.iloc[test])
    y_ac_pred = list(map(list,zip(y_actual,y_pred)))
    total = len(y_ac_pred)
    count = 0
    
    # for rows in y_ac_pred:
    #   if rows[0] == rows[1]:
    #     count += 1

    cnt_act = Counter(y_actual)
    cnt_pred = Counter(y_pred)
    # print(y_ac_pred)
    print(metrics.accuracy_score(y_actual, y_pred))
    print(cnt_act)
    print(cnt_pred)
    # print(count, total, count / total)

    confusion_matrix = metrics.confusion_matrix(y_actual, y_pred)
    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix,display_labels = ['a','c','p'] ) #

    cm_display.plot()
    plt.show()





  # print(numpy.mean(scores))
  # print(scores)
  
```
